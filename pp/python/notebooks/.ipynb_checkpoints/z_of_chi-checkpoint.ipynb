```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
# import camb
from astropy.cosmology import Planck18,z_at_value
from IPython.display import display, Markdown # for displaying python output in markdown
plt.rcParams['text.usetex'] = True
```

```python
##############################
###   Physical constants   ###
##############################

# speed of light (km/s)
c = 2.99792458e5 * u.km/u.s

###################################
###   Cosmological parameters   ###
###################################
# following Planck 2018 results

# reduced Hubble constant
h = 0.6735

# Hubble constant (km/s/Mpc)
H_0 = 100*h * u.km/u.s/u.Mpc

# baryon density fraction
Omega_b = 0.0493

# DM density fraction
Omega_X = 0.2645

# matter density fraction
Omega_m = Omega_b + Omega_X
```

Given a comoving distane $\chi$ (in Mpc), we can calculate the corresponding redshift $z$

```python
# given a comoving distance (in Mpc)
chi = 10000. * u.Mpc

# the redshift is
z = z_at_value( Planck18.comoving_distance, chi )

dist_u = 'Mpc'
if chi.value>=1000.:
    chi=chi.to(u.Gpc)
    dist_u = 'Gpc'
elif chi.value<1.:
    chi=chi.to(u.kpc)
    dist_u = 'kpc'

display(Markdown( r'An object at comoving distance $\chi={:0.5} ~ \text{{{}}}$ has redshift $z={:2.5}$.'.format( chi.value, dist_u, z.value ) ))
```

and simliarly, given a redshift $z$, we can calculate the corresponding comoving distnace $\chi$

```python
# given a redshift
z = 1.

# the comoving distnace (in Mpc) is
chi = Planck18.comoving_distance(z)
dist_u = 'Mpc'
if chi.value>=1000.:
    chi=chi.to(u.Gpc)
    dist_u = 'Gpc'
elif chi.value<1.:
    chi=chi.to(u.kpc)
    dist_u = 'kpc'

display(Markdown( r'Objects at redshift $z={:0.5}$ are at comoving distance $\chi={:0.5} ~ \text{{{}}}$.'.format( z, chi.value, dist_u ) ))
```

Note that scale factor $a$ and redshift $z$ are related by $a^{-1}=1+z$.

```python
def z2a(z): return 1/(1+z)
def a2z(a): return 1/a - 1

chi = np.linspace(1*u.Mpc,1.2e4*u.Mpc,1000)
z   = z_at_value( Planck18.comoving_distance, chi, zmax=1.e6 )

fig,ax = plt.subplots(1,2)
ax[0].plot(chi,z)
ax[0].set_xlabel( r'comoving distance $\chi$ (Mpc)' )
ax[0].set_ylabel( r'redshift $z$' )

ax[1].plot(chi,z)
ax[1].set_xlabel( r'comoving distance $\chi$ (Mpc)' )
ax[1].set_ylabel( r'redshift $z$' )
ax[1].set_xscale('log')
ax[1].set_yscale('log')

ax2 = ax[1].twinx()
ax2.set_yscale('log')
ax2.set_yticks( ax[1].get_yticks() )
ax2.set_ybound(ax[1].get_ybound())
ax2.set_yticklabels([round(z2a(y),3) for y in ax[1].get_yticks()])
ax2.set_ylabel( r'scale factor $a$' )

# secyax = ax[1].secondary_yaxis('right', functions=(z2a, a2z))
# secyax.set_ylabel('scale factor $a$')

fig.set_size_inches(12, 5, forward=True)

plt.show()
```

Compare to Planck 2018 cosmology

```python
from astropy.cosmology import FlatLambdaCDM

# New Model Parameters
Omega_x_new = .26600
Omega_b_new = .0493
Omega_Lambda_new = 1 - Omega_x_new - Omega_b_new
h_new = .6735
ns_new = .9649
sigma8_new = .81147

cosmo = FlatLambdaCDM(H0=h_new*100, Om0=Omega_b_new+Omega_x_new, Ob0=Omega_b_new)


chi = np.linspace(1*u.Mpc,1.2e4*u.Mpc,1000)
z   = z_at_value( Planck18.comoving_distance, chi, zmax=1.e6 )
z_new = z_at_value( cosmo.comoving_distance, chi, zmax=1.e6 )

fig,ax = plt.subplots(1,2)
ax[0].plot(chi,z)
ax[0].plot(chi,z_new)
ax[0].set_xlabel( r'comoving distance $\chi$ (Mpc)' )
ax[0].set_ylabel( r'redshift $z$' )

ax[1].plot(chi,z)
ax[1].plot(chi,z_new)
ax[1].set_xlabel( r'comoving distance $\chi$ (Mpc)' )
ax[1].set_ylabel( r'redshift $z$' )
ax[1].set_xscale('log')
ax[1].set_yscale('log')

ax2 = ax[1].twinx()
ax2.set_yscale('log')
ax2.set_yticks( ax[1].get_yticks() )
ax2.set_ybound(ax[1].get_ybound())
ax2.set_yticklabels([round(z2a(y),3) for y in ax[1].get_yticks()])
ax2.set_ylabel( r'scale factor $a$' )

# secyax = ax[1].secondary_yaxis('right', functions=(z2a, a2z))
# secyax.set_ylabel('scale factor $a$')

fig.set_size_inches(12, 5, forward=True)

plt.show()
```
